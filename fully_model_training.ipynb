# 一个完整的模型训练套路

```python
from PIL import Image
import torchvision
import torch.nn as nn 
import torch
```

```python
img_path = "images/dog.png" 
img = Image.open(img_path)
img = img.convert("RGB") # 因为png是四通道，除了RGB三个通道外，还有一个 透明度 通道
```

```python
transform = torchvision.transforms.Compose([torchvision.transforms.Resize((32,32)),torchvision.transforms.ToTensor()])
img  =transform(img)
```

```python
print(img.shape)
```

```python
class Tudui(nn.Module):
    def __init__(self):
        super(Tudui, self).__init__()
        self.model = nn.Sequential(
            nn.Conv2d(3, 32, 5, 1, 2),
            nn.MaxPool2d(2),
            nn.Conv2d(32, 32, 5, 1, 2),
            nn.MaxPool2d(2),
            nn.Conv2d(32, 64, 5, 1, 2),
            nn.MaxPool2d(2),
            nn.Flatten(),
            nn.Linear(64*4*4, 64),
            nn.Linear(64, 10)
        )
    def forward(self, x):
        x = self.model(x)
        return x
```

```python
model = torch.load("models/tudui_10_epoch.pth", weights_only=False)
print(model)
```

```python
img = torch.reshape(img, (1, 3, 32, 32))
model.eval()
with torch.no_grad():
    output = model(img)
print(output)
print(output.argmax(1)) # 第四个类别
```
