```python
import torch
```

张量表示一个数值组成的数组，这个数组可能有多个维度

```python
x = torch.arange(12)
x
```

通过张量的 ·shape· 属性来访问张量的形状和张量中元素的总数

```python
x.shape
```

```python
x.numel()
```

- 改变张量的形状，但是不改变他的元素数量和元素值，可以调用 `reshape` 函数

```python
X = x.reshape(3, 4)
X
```

```python
import torch 

a = torch.tensor([[1,2],[3,4]])

a.type()

a = torch.tensor((2,3))
print(a.type())
a.size()

torch.ones((3,3), dtype=int)
torch.eye(2,dtype=int)


''' 随机 '''
a = torch.rand((2,3))
print(a)

b = torch.normal(mean=0, std=torch.rand(5))
print(b)

c = torch.Tensor(2,2).uniform_(-1,1)  # 均匀分布
print(c)


''' 序列 '''
d = torch.arange(0,11,2) # 起点，终点，步长
print(d)

e = torch.linspace(2,10,4) # 拿到等间隔的4个数字
print(e)

f = torch.randperm(10) # 生成从0-9的随机序列
print(f)
```
