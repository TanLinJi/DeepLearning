### Tensor的属性



1. 稀疏的张量，可以使得模型变得更简单
2. 减少内存开销

```python
# 使用张量的属性
import torch

dev = torch.device("cuda:0") # 写 'cuda' 会使用默认的显卡
a = torch.tensor([2,2], dtype=torch.float32, device=dev)
print(a)
```

```python
# 定义一个稀疏的张量
# 如果不去指定tensor的属性，默认都是使用的是稠密的张量 
x = torch.tensor([[0,1,2],[0,1,2]]) # 坐标(0,0)(1,1)(2,2)
v = torch.tensor([1,2,3])  # 值
a = torch.sparse_coo_tensor(x,v,(4,4)) # 4*4的size
print(a)
b = torch.sparse_coo_tensor(x,v,(4,4)).to_dense() # 转成稠密的张量
print(b)

c = torch.sparse_coo_tensor(x,v,(4,4),device=dev, dtype=torch.float32).to_dense() # 修改属性，放在GPU上
print(c)
```

### Tnsor的算术运算

```python
c = a + b
c = torch.add(a,b)
a.add(b)
a.add_(b)  # 注意：所有后边这种带下划线的，都会在计算结果的同时，对原来a的值进行修改
```

```python
c = a - b
c = torch.sub(a,b)
a.sub(b)
a.sub_(b)  # 注意：所有后边这种带下划线的，都会在计算结果的同时，对原来a的值进行修改
```
