# Loss Function

1. 计算实际输出和目标之间的差距
2. 为我们更新输出提供一定的依据（反向传播）

### L1Loss && MSELoss

```python
import torch

inputs = torch.tensor([1,2,3],dtype=torch.float)
targets = torch.tensor([1,2,5],dtype=torch.float)

inputs = torch.reshape(inputs, (1,1,1,3))  # 代表batchsize=1, channel=1, row=1, col=1

targets = torch.reshape(targets, (1,1,1,3))

loss_l1 = torch.nn.L1Loss()
result_l1 = loss_l1(inputs, targets)

print(result_l1)

loss_mse = torch.nn.MSELoss() # https://docs.pytorch.org/docs/stable/generated/torch.nn.MSELoss.html#torch.nn.MSELoss
result_mse = loss_mse(inputs, targets) 
print(result_mse)
```

### CrossEntropyLoss

It is useful when training a classification problem with C classes. If provided, the optional argument weight should be a 1D Tensor assigning weight to each of the classes. This is particularly useful when you have an unbalanced training set.(也就是说对分类问题比较适用)

```python
import torch.nn as nn 
import torch

x = torch.tensor([0.1, 0.2, 0.3])
y = torch.tensor([1])
x  =torch.reshape(x, (1, 3))

loss_cross = nn.CrossEntropyLoss()
result_cross = loss_cross(x,y)  # -0.2 + ln(exp(0.1) + exp(0.2) + exp(0.3))
print(result_cross)
```
