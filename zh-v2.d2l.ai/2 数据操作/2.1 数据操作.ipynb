```python
import torch
```

```python
# 1. 使用arrange创建一个行向量
x = torch.arange(12)  # 默认创建为整数,另外，除非额外指定，新创建好的张量都默认存储在内存中，并在CPU上计算
x
```

```python
# 2. 通过张量的shape属性来访问张量的形状
x.shape
```

```python
# 3. 通过numel()来获取张量中元素的个数
x.numel()
```

```python
# 4. 调用reshape函数，来改变张量的形状，但是不改变元素数量和值
X = x.reshape(3,4)
X
```

```python
# 5. 上边我们指定了是3行4列
#    如果，我们只关心3行，剩下一个维度的值可以填-1，这样他会自动计算出另外一个维度
X1 = x.reshape(3,-1) # 如果这个维度除不尽会报错
X1
```

```python
# 6. 获取一个全0张量
torch.zeros((2,3,4))
```

```python
# 7. 获取一个全1矩阵
torch.ones((2,3))
```

```python
# 8. 从特定的概率分布中随机采样得到张量中每个元素的值
# 从均值为0，标准差为1的标准正态分布中采样
torch.randn(3,4)
```

```python
# 9. 通过提供包含数值的Python列表（或嵌套列表），来为所需张量中的每个元素赋予确定值
torch.tensor(
    [[2, 1, 4, 3], 
    [1, 2, 3, 4], 
    [4, 3, 2, 1]]
    )
```
