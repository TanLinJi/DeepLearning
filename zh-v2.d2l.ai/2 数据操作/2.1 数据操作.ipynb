### 2.1.2. 运算符

对于任意具有相同形状的张量， 常见的标准算术运算符（+、-、*、/和**）都可以被升级为按元素运算，可以在同一形状的任意两个张量上调用按元素操作。

```python
x = torch.tensor([1.0, 2, 4, 8])
y = torch.tensor([2, 2, 2, 2])
x + y, x - y, x*y, x/y, x**y
```

```python
torch.exp(torch.Tensor([0])) # 求e的x次方
```

```python
# “按元素”方式可以应用更多的计算，包括像求幂这样的一元运算符。
torch.exp(x)
```

### 2.1.3. 广播机制

对于形状不同的两个张量，也可以通过调用广播机制来执行按孙苏操作。这种机制的工作方式如下：<br>
1. 通过适当复制元素来扩展一个或两个数组，以便在转换后，两个张量有相同的形状
2. 对生成的元素执行按元素操作

广播机制规则：<br>

- 如果遵守以下规则，则两个tensor是“可广播的”
    - 每个tensor至少有一个维度；
    - 遍历tensor所有维度时，从末尾随开始遍历，两个tensor存在下列情况：
        - tensor维度相等。
        - tensor维度不等且其中一个维度为1。
        - tensor维度不等且其中一个维度不存在。
- 如果两个tensor是“可广播的”，则计算过程遵循下列规则：
    - 如果两个tensor的维度不同，则在维度较小的tensor的前面增加维度，使它们维度相等。
    - 对于每个维度，计算结果的维度值取两个tensor中较大的那个值。
    - 两个tensor扩展维度的过程是将数值进行复制。

```python
# 1. 在大多数情况下，沿着数组中长度为1的轴进行广播
a = torch.arange(3).reshape((3,1))
b = torch.arange(2).reshape((1,2))
a,b
```

```python
# 由于a和b分别是3*1和1*2矩阵，如果让它们相加，它们的形状不匹配。 
# 将两个矩阵广播为一个更大的矩阵，如下所示：矩阵a将复制列， 矩阵b将复制行，然后再按元素相加。

a + b
```

### 2.1.5. 节省内存
运行一些操作可能会导致为新结果分配内存。 例如，Y = X + Y，将取消引用Y指向的张量，而是指向新分配的内存处的张量。

```python
# 1. 使用id()函数来查看内存中引用对象的地址
before = id(Y)
Y = Y + X  # 这个操作是：Python首先计算Y + X，为结果分配新的内存，然后使Y指向内存中的这个新位置。
now = id(Y)
before, now, now==before
```

可能这样做不是总是可取的，原因有：<br>
1. 我们不想总是不必要地分配内存，而是希望原地更新这些参数
2. 如果我们不原地更新这些参数，其他引用依然会指向旧的内存位置，可能导致某些代码会无意当中引用到旧的没有更新的参数

```python
# 2. 使用切片表示法讲操作的结果分配给先前分配的数组，例如Y[:]=...
Z = torch.zeros_like(Y)  # 创建一个和Y形状相同的全0矩阵
print('ID(z)：',id(Z))
Z[:] = X + Y
print('ID(z)：',id(Z))
```

```python
# 3. 如果在后续计算中没有重复使用X， 也可以使用X[:] = X + Y或X += Y来减少操作的内存开销。
before = id(X)
X += Y
id(X) == before
```
